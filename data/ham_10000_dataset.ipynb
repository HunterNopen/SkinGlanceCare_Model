```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ham_df = pd.read_csv("datasets/HAM10000_metadata.csv")
```

```python
class_counts = ham_df['dx'].value_counts()
print(class_counts)
print(f"\nTotal: {len(ham_df)}")
```

```python
plt.figure(figsize=(10, 6))
class_counts.plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Class Distribution')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
